```python
import pandas as pd
```

```python
import requests 

docs_url = 'https://github.com/alexeygrigorev/llm-rag-workshop/raw/main/notebooks/documents.json'
docs_response = requests.get(docs_url)
documents_raw = docs_response.json()

documents = []

for course in documents_raw:
    course_name = course['course']

    for doc in course['documents']:
        doc['course'] = course_name
        documents.append(doc)
```

```python
# inspecting data - getting the first item from the doc list
df = documents[0]
```

```python
# creating the actual pandas dataframe from the full list, with specified order
df = pd.DataFrame(documents, columns=['course','section','question','text'])
```

```python
# df.head()
# df.tail()
# filtering in specific course
df[df.course == 'data-engineering-zoomcamp']
```

### Text Search

#### Count Vectorizer 

```python
# turning text to vectors
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(min_df=5)

# earns the vocabulary from the doc
cv.fit(df.text)

# Extracts the list of words (aka features or columns) from the vocabulary
cv.get_feature_names_out()
```

```python
doc_examples = [
    "Course starts on 15th Jan 2024",
    "Prerequisites listed on GitHub",
    "Submit homeworks after start date",
    "Registration not required for participation",
    "Setup Google Cloud and Python before course"
]
cv = CountVectorizer(stop_words='english')
cv.fit(docs_examples)
cv.get_feature_names_out()
```

```python
X = cv.transform(doc_examples)
pd.DataFrame(X.todense(),columns=cv.get_feature_names_out()).T
```

#### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(stop_words='english', min_df=5)
X = cv.fit_transform(df.text)

names = cv.get_feature_names_out()

df_docs = pd.DataFrame(X.toarray(), columns=names)
df_docs.round(2)
```

#### Query-Document Similarity

```python
query = "Do I need to know python to sign up for the January course?"

q = cv.transform([query])
q.toarray()
```

```python
# convert the already vectorized query into a key-value pair dictionary (readable)
query_dict = dict(zip(names, q.toarray()[0]))
query_dict

doc_dict = dict(zip(names, X.toarray()[1]))
doc_dict
```
